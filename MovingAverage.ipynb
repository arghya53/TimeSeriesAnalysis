Moving Average is basically used in case of stationary data. If the data has non-stationary mean then it is not a perfect method for forecasting. Here, we are going to see the performance of moving averages when applied on non-stationary data.

```python
# indexing the datetime data to make partial or sub string search easier
bill_df = df.set_index('Bill_Date')
```

```python
bill_df.head()
```

```python
bill_2018 = bill_df['2016':'2018'][['Billed_amount']]
bill_2018
```

**Simple Moving Average** = (t+(t-1)+(t-2)+.....+(t-n))/n

```python
# rolling window=3, calculates the mean of the last three values in the 'Billed_amount' column for each row.
bill_2018['Billed_amount'].rolling(window=3).mean()
```

```python
# Shift(1) essentially shifts the rolling mean values one row upward. This is often done to align the rolling mean with the corresponding data point.
bill_2018['ma_rolling'] = bill_2018['Billed_amount'].rolling(window=3).mean().shift(1)
bill_2018
```

```python
bill_2018.plot()
```

**Weighted Moving Avg**= (t* weighting factor)+((t-1)* weighting factor-1) + ((t-n)*weighting factor-n)/n

```python
def wma(weights):
  def calc(x):
    return (weights*x).mean()
  return calc
```

```python
bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5])))
```

```python
bill_2018['wma_rolling']=bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
```

```python
bill_2018
```

```python
bill_2018.plot()
```

**Exponential Moving Average** = (Close-previous EMA)*(2/(span+1)) + previous EMA

```python
bill_2018['ewm_window_3']=bill_2018['Billed_amount'].ewm(span=3, adjust=False, min_periods=0).mean().shift(1)
```

```python
bill_2018
```

```python
bill_2018.plot()
```

**Exponential Smoothing Average** Y(t+1) = @[Y(t)+(1-@)* Y(t-1)+(1-@)^2* Y(t-2)+....]

```python
# Here, alpha is the smoothing factor which controls the exponential decay.
# Alpha increases to weight more on present data and decreases to weight more on past data
bill_2018['esm_window_3_7']=bill_2018['Billed_amount'].ewm(alpha=0.7, adjust=False, min_periods=3).mean().shift(1)
```

```python
bill_2018
```

```python
bill_2018.plot()
```

```python
bill_2018[['Billed_amount', 'esm_window_3_7']].plot()
```

```python
bill_2018['esm_window_3_3']=bill_2018['Billed_amount'].ewm(alpha=0.3, adjust=False, min_periods=3).mean().shift(1)
```

```python
bill_2018
```

```python
bill_2018[['Billed_amount', 'esm_window_3_3']].plot() # here we see that, as the alpha is reduced, the curve becomes smooth, but it can't calculate the fluctuations properly
```

**root mean squared error**  image_2023-10-17_163346958.png

```python
((bill_2018['Billed_amount']-bill_2018['ma_rolling'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['wma_rolling'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['ewm_window_3'])**2).mean()**0.5
```

```python
# We can see that the squared error is less in case of exponential smoothing avg with smoothing factor 0.7
((bill_2018['Billed_amount']-bill_2018['esm_window_3_7'])**2).mean()**0.5
```

```python
((bill_2018['Billed_amount']-bill_2018['esm_window_3_3'])**2).mean()**0.5
```
